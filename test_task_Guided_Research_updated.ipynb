```python
import torch
import torch.nn as nn
import torch.distributions as dist

# defining architecture of the network
input_data, hidden_layers_fc1,hidden_layers_fc2 ,output_size, batch_size, num_classes,dist_value = 100, 5,10, 1, 10, 1,0
#dist_value = 0

# Create dummy input and target tensors (data)
x = torch.randn(batch_size, input_data)
y = torch.tensor([[1.0], [0.0], [0.0], [1.0], [1.0], [1.0], [0.0], [0.0], [1.0], [1.0]])


# model creation
#ctrl + / for multiline comments
# model = nn.Sequential(nn.Linear(input_data, hidden_layers),
#                      nn.ReLU(),
#nn.Linear(hidden_layers, output_size))


#randomly sampling from the categorical data 
m = dist.Categorical(torch.tensor([0.5,0.5]))
# #print (m)
# sample_value = m.sample()
#print (sample_value.item())

class NeuralNet(nn.Module):
    #model creation
    def __init__(self, input_data, hidden_layers_fc1,hidden_layers_fc2, num_classes,dist_value):
        super(NeuralNet, self).__init__()
        self.fc1 = nn.Linear(input_data, hidden_layers_fc1) 
        self.relu = nn.ReLU()
        self.fc2 = nn.Linear(input_data, hidden_layers_fc2)  
        self.final_fc1 = nn.Linear(hidden_layers_fc1,num_classes)
        self.final_fc2 = nn.Linear(hidden_layers_fc2,num_classes)
        self.sigmod = nn.Sigmoid()
    #forward pass
    def forward(self, x,dist_value):
        #select which linear layer to use in the forward pass
        if(dist_value == 1):
            
            out = self.fc1(x)
            out = self.relu(out)
            out = self.final_fc1(out)
            #out = self.sigmod(out)
            print('selection 1')
            return out
        elif(dist_value == 0):
            out = self.fc2(x)
            out = self.relu(out)
            out = self.final_fc2(out)
            #out = self.sigmod(out)
            print('selection 0')
            return out
    
model = NeuralNet(input_data, hidden_layers_fc1,hidden_layers_fc2 , num_classes,dist_value)


# loss function
criterion = nn.MSELoss()


# optimizer for the network
optimizer = torch.optim.SGD(model.parameters(), lr=0.01, momentum = 0.05)

var_loss = []
reward = 0
# Gradient Descent
for epoch in range(500):
    #picking samples from the categorical distrbution
    
    sample_value = m.sample()
    # Predicted value  
    y_pred = model(x,sample_value.item())
    #print (y_pred)
    # Loss calculation
    loss = criterion(y_pred, y)
    #print('epoch: ', epoch,' loss: ', loss.item())
    
    #Reward; Punishment functions
    if (epoch == 0):
        var_loss.append(loss.item())
        
        
    elif (var_loss[epoch-1] > loss.item()):
        print ('--Reward Section--')
        print ('last loss',var_loss[epoch-1])
        print ('current loss', loss.item())
        # reward logic
        loss = (loss*0.5)
        reward = 0.5
       # loss = -m.log_prob(sample_value) * 0.5
       # print (loss.item())
        var_loss.append(loss.item())
        #print (fin_loss.item())
    else:
        print ('--Punishment Section--') 
        #punishment logic
        loss = loss*1.25
        reward = 1.25
        #loss = -m.log_prob(sample_value) * 1.5
        print ('last loss',var_loss[epoch-1])
        print ('current loss', loss.item())
        var_loss.append(loss.item())
        #print(loss.item())
        
    optimizer.zero_grad()
    #backpropogation on the loss
    loss.backward()
    #calculate the score function here based on the categorical distribution
    score = -m.log_prob(sample_value) * reward
    print ("score for this epoch is:",score.item())
    # parameters update
    optimizer.step()
#print (var_loss)
```
